```python
# importing libraries
import pandas as pd
import numpy as np
```

```python
# load ratings data
ratings = pd.read_csv("ratings.csv")
```

```python
# load tags data
tags = pd.read_csv("tags.csv")
```

```python
# drop timestamp (not needed)
tags = tags.drop(columns=['timestamp'])
```

```python
tags.columns
```

```python
# select required columns
data = tags[['userId', 'movieId', 'tag']]
```

```python
#STEP 1: Clean the tags 
data['tag'] = data['tag'].str.lower()
```

```python
#STEP 2: Combine all tags for each movie
movie_tags = data.groupby('movieId')['tag'].apply(lambda x: ' '.join(x))
movie_tags.head()
```

```python
#STEP 3: Convert tags into vectors
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()
movie_tfidf = tfidf.fit_transform(movie_tags)
```

```python
#STEP 4: Calculate similarity between movies
from sklearn.metrics.pairwise import cosine_similarity
movie_similarity = cosine_similarity(movie_tfidf)
```

```python
# STEP 5: Convert similarity into DataFrame
movie_similarity_df = pd.DataFrame(movie_similarity,index=movie_tags.index,columns=movie_tags.index)
```

```python
#STEP 6: Checking the existing Movie Ids
movie_similarity_df.columns[:10]
```

```python
InputMovie = 5
similar_movies = movie_similarity_df[InputMovie].sort_values(ascending=False)
```
